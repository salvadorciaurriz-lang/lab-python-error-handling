# 2. Función `initialize_inventory`

## Objetivo

Crear un inventario preguntando al usuario cuántas unidades hay disponibles de cada producto.

## Manejo de errores

La función controla:

- valores no numéricos
- cantidades negativas
- entradas vacías
- repite la pregunta hasta que la cantidad sea válida

```python
def initialize_inventory(products):
    inventory = {}

    for product in products:
        while True:
            try:
                quantity = input(f"Enter the quantity of {product}s available: ")

                if quantity.strip() == "":
                    raise ValueError("Input cannot be empty.")

                quantity = int(quantity)

                if quantity < 0:
                    raise ValueError("Quantity cannot be negative.")

            except ValueError as error:
                print(f"Error: {error}. Please enter a valid non-negative integer.")

            else:
                inventory[product] = quantity
                break

            finally:
                # finally always runs, whether there is an error or not.
                # In this case, we keep it simple to demonstrate the structure.
                pass

    return inventory
```

# 3. Función `get_customer_orders`

## Objetivo

Pedir al usuario cuántos productos quiere pedir y qué productos desea.

## Manejo de errores

La función controla:

- número de pedidos no numérico
- número de pedidos negativo
- productos que no existen en el inventario
- productos sin stock
- entradas vacías

La función devuelve un `set` con los productos pedidos.

```python
def get_customer_orders(inventory):
    while True:
        try:
            number_of_orders = input("Enter the number of customer orders: ")

            if number_of_orders.strip() == "":
                raise ValueError("Input cannot be empty.")

            number_of_orders = int(number_of_orders)

            if number_of_orders < 0:
                raise ValueError("Number of orders cannot be negative.")

        except ValueError as error:
            print(f"Error: {error}. Please enter a valid non-negative integer.")

        else:
            break

    customer_orders = set()

    for _ in range(number_of_orders):
        while True:
            try:
                product = input("Enter the name of a product that a customer wants to order: ").lower().strip()

                if product == "":
                    raise ValueError("Product name cannot be empty.")

                if product not in inventory:
                    raise KeyError(f"'{product}' is not available in the inventory.")

                if inventory[product] <= 0:
                    raise ValueError(f"'{product}' is out of stock.")

            except KeyError as error:
                print(f"Error: {error}. Please choose a product from the inventory.")

            except ValueError as error:
                print(f"Error: {error}. Please choose another product.")

            else:
                customer_orders.add(product)
                break

    return customer_orders
```

# 4. Función `update_inventory`

## Objetivo

Actualizar el inventario restando una unidad por cada producto pedido.

## Manejo de errores

La función controla:

- productos no encontrados
- productos sin stock
- inventarios con cantidades inválidas

```python
def update_inventory(customer_orders, inventory):
    try:
        if not isinstance(customer_orders, set):
            raise TypeError("customer_orders must be a set.")

        if not isinstance(inventory, dict):
            raise TypeError("inventory must be a dictionary.")

        for product in customer_orders:
            if product not in inventory:
                raise KeyError(f"'{product}' does not exist in the inventory.")

            if not isinstance(inventory[product], int):
                raise TypeError(f"The quantity for '{product}' must be an integer.")

            if inventory[product] <= 0:
                raise ValueError(f"'{product}' is out of stock and cannot be ordered.")

            inventory[product] -= 1

    except (KeyError, TypeError, ValueError) as error:
        print(f"Error updating inventory: {error}")

    else:
        print("Inventory updated successfully.")

    finally:
        return inventory
```

# 5. Función `calculate_order_statistics`

## Objetivo

Calcular estadísticas del pedido:

- número total de productos pedidos
- porcentaje de productos únicos pedidos respecto al catálogo completo

## Manejo de errores

La función controla:

- que `customer_orders` sea un set
- que `products` sea una lista
- que la lista de productos no esté vacía

```python
def calculate_order_statistics(customer_orders, products):
    try:
        if not isinstance(customer_orders, set):
            raise TypeError("customer_orders must be a set.")

        if not isinstance(products, list):
            raise TypeError("products must be a list.")

        if len(products) == 0:
            raise ValueError("products list cannot be empty.")

        total_products_ordered = len(customer_orders)
        percentage_ordered = (total_products_ordered / len(products)) * 100

    except (TypeError, ValueError) as error:
        print(f"Error calculating order statistics: {error}")
        return None

    else:
        return total_products_ordered, percentage_ordered
```

# 6. Función `print_order_statistics`

## Objetivo

Mostrar las estadísticas del pedido de forma clara.

## Manejo de errores

La función controla:

- que las estadísticas no sean `None`
- que tengan el formato correcto

```python
def print_order_statistics(order_statistics):
    try:
        if order_statistics is None:
            raise ValueError("Order statistics are missing.")

        if not isinstance(order_statistics, tuple):
            raise TypeError("Order statistics must be a tuple.")

        if len(order_statistics) != 2:
            raise ValueError("Order statistics must contain exactly two values.")

        total_products_ordered, percentage_ordered = order_statistics

    except (TypeError, ValueError) as error:
        print(f"Error printing statistics: {error}")

    else:
        print("Order Statistics:")
        print(f"Total Products Ordered: {total_products_ordered}")
        print(f"Percentage of Products Ordered: {percentage_ordered:.2f}%")
```

# 7. Función `calculate_total_price`

## Objetivo

Calcular el precio total del pedido usando un diccionario de precios.

## Manejo de errores

La función controla:

- que los pedidos existan en el diccionario de precios
- que los precios sean numéricos
- que los precios no sean negativos

```python
def calculate_total_price(customer_orders, prices):
    total_price = 0

    try:
        if not isinstance(customer_orders, set):
            raise TypeError("customer_orders must be a set.")

        if not isinstance(prices, dict):
            raise TypeError("prices must be a dictionary.")

        for product in customer_orders:
            if product not in prices:
                raise KeyError(f"'{product}' does not have a price assigned.")

            price = prices[product]

            if not isinstance(price, (int, float)):
                raise TypeError(f"The price for '{product}' must be numeric.")

            if price < 0:
                raise ValueError(f"The price for '{product}' cannot be negative.")

            total_price += price

    except (KeyError, TypeError, ValueError) as error:
        print(f"Error calculating total price: {error}")
        return None

    else:
        return total_price
```

# 8. Función principal `main`

## Objetivo

Integrar todo el flujo:

1. Crear inventario.
2. Pedir productos al cliente.
3. Actualizar inventario.
4. Calcular estadísticas.
5. Calcular precio total.
6. Mostrar resultados.

```python
def main():
    products = ["t-shirt", "mug", "hat", "book", "keychain"]

    prices = {
        "t-shirt": 15.0,
        "mug": 8.0,
        "hat": 12.0,
        "book": 10.0,
        "keychain": 5.0
    }

    inventory = initialize_inventory(products)

    print("\nCurrent inventory:")
    print(inventory)

    customer_orders = get_customer_orders(inventory)

    print("\nCustomer orders:")
    print(customer_orders)

    updated_inventory = update_inventory(customer_orders, inventory)

    print("\nUpdated inventory:")
    print(updated_inventory)

    order_statistics = calculate_order_statistics(customer_orders, products)

    print()
    print_order_statistics(order_statistics)

    total_price = calculate_total_price(customer_orders, prices)

    if total_price is not None:
        print(f"Total price: {total_price:.2f} €")
```
